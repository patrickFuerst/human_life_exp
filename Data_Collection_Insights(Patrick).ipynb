### Load World by Income 

The world bank provides sofisticated online tools to collect the data you are interested in. 
Still I didn't found a way to export the data by Income-Level per country in a nice formart
http://datatopics.worldbank.org/world-development-indicators/the-world-by-income-and-region.html
If we wanna download the data from this link it throws an error. 

The Income-Level is defineded based on the GNI (Gross National Income) and the income group threshold is determined as descibred here. https://datahelpdesk.worldbank.org/knowledgebase/articles/378833-how-are-the-income-group-thresholds-determined
We are downloading all the country GNI data [here](https://data.worldbank.org/indicator/NY.GNP.PCAP.CD?end=2017&locations=XM-AF-ET&start=2017&view=map). 
__The levels (high, low e.g) provided in the country data are not the threshold levels as defined here!__ https://datahelpdesk.worldbank.org/knowledgebase/articles/906519 


__Since we didn't find the threshold values in the data bank, we extract the data from the Excel files found here:__
https://datahelpdesk.worldbank.org/knowledgebase/articles/906519-world-bank-country-and-lending-groups
__See also `data/raw` folder.__

The levels are defined by the following: 
* Low income (L)
* Lower middle income (LM)
* Upper middle income (UM)
* High income (H)

# Preprocess data 

Let's preprocess it so we can merge it later.
Let's also look how many years we have for Life Expectancy, so we can remove all other years of the other data. 


### Process Life Expectancy

We are interested in the `Life expectancy at birth` and thus remove all other columns. 
Thus we also could have a look at HALE (Healthy life expectancy), for a lot of countries there is very sparse data available.


```python
df_life_exp.columns.rename([None,None], inplace=True)
df_life_exp.index.names = (['Country', 'Year'])
```

```python

df_life_exp.drop(df_life_exp.columns[3:], axis=1,  inplace=True)
display(df_life_exp.head(5))
```

```python
# be sure the types are right 
#we get an error when converting to float, since the country Rwanda contains two values per colunm. 
#we need to filter this out
#we decide that we take just the first one

def filter_values(v):
    return v.split(' ')[0]

df_life_exp.iloc[:,0] = df_life_exp.iloc[:,0].apply(filter_values )
df_life_exp.iloc[:,1] = df_life_exp.iloc[:,1].apply(filter_values )
df_life_exp.iloc[:,2] = df_life_exp.iloc[:,2].apply(filter_values )
```

```python
df_life_exp = df_life_exp.apply(pd.to_numeric)
display(df_life_exp.dtypes)
display(df_life_exp.head(5))
```

```python
display(df_life_exp.describe())
display('Lowest Year ' + str(df_life_exp.index.get_level_values(1).min()))
```

### Process Current health expenditure

```python
df_health_expenditure = df_health_expenditure.apply(pd.to_numeric)
display(df_health_expenditure.dtypes)
display(df_health_expenditure.head(5))
```

```python
display('Lowest Year ' + str(df_health_expenditure.index.get_level_values(1).min()))
```

### Process World by Income 


```python
display(df_world_income.head(5))
```

```python
df_world_income.replace('..', 'Unknown', inplace=True)
display(df_world_income.head(5))
```

```python
df_world_income.drop(df_world_income.columns[0], axis=1, inplace=True)
df_world_income.index.set_names(['Country'],inplace=True)
df_world_income.columns.names= ['Year']
```

```python
df_world_income = pd.DataFrame(df_world_income.stack())
```

```python
df_world_income.columns = ['Income Level']
```

```python
display(df_world_income.head(5))
df_world_income.iloc[:,0].unique() # There is a LM* defined wich is specific to Yemen. We gonna remove this anyway by the year later
```

```python
df_world_income_filtered = df_world_income.reset_index()
df_world_income_filtered = df_world_income_filtered[df_world_income_filtered['Year'].astype(float) >= 2000]
df_world_income_filtered.set_index(['Country','Year'], inplace=True)
df_world_income = df_world_income_filtered
display(df_world_income_filtered.head(5))
df_world_income.iloc[:,0].unique()
```

### Process GDP per capita


```python
display(df_gpd_per_capita.head(5))
```

```python
df_gpd_per_capita.drop(['Series Code','Series Name','Country Code'], axis=1, inplace=True)
```

```python
def truncate_column(c):
    return c.split(' ')[0]

df_gpd_per_capita = df_gpd_per_capita.rename(truncate_column,axis=1)
```

```python
df_gpd_per_capita_filtered = df_gpd_per_capita.set_index('Country')
df_gpd_per_capita_filtered.columns.names= ['Year']
df_gpd_per_capita_filtered = pd.DataFrame(df_gpd_per_capita_filtered.stack())
df_gpd_per_capita_filtered = df_gpd_per_capita_filtered.reset_index()
df_gpd_per_capita_filtered = df_gpd_per_capita_filtered[df_gpd_per_capita_filtered['Year'].astype(float) >= 2000]
df_gpd_per_capita_filtered.set_index(['Country','Year'], inplace=True)
```

```python
df_gpd_per_capita = df_gpd_per_capita_filtered
df_gpd_per_capita.columns = ['GDP per Capita']
display(df_gpd_per_capita.head(5))
```

```python
display(df_gpd_per_capita.head(5))
display(df_gpd_per_capita.describe())
```

### Process Research and development expenditure

```python
df_research_dev_exp.drop(['Series Code','Series Name','Country Code'], axis=1, inplace=True)
df_research_dev_exp = df_research_dev_exp.rename(truncate_column,axis=1)
```

```python
df_research_dev_exp_filtered = df_research_dev_exp.set_index('Country')
df_research_dev_exp_filtered.columns.names= ['Year']
df_research_dev_exp_filtered = pd.DataFrame(df_research_dev_exp_filtered.stack())
df_research_dev_exp_filtered = df_research_dev_exp_filtered.reset_index()
df_research_dev_exp_filtered = df_research_dev_exp_filtered[df_research_dev_exp_filtered['Year'].astype(float) >= 2000]
df_research_dev_exp_filtered.set_index(['Country','Year'], inplace=True)
```
